```python
import pandas as pd
import pickle

# Load test data
test_data = pd.read_csv('/Users/masato/Documents/GitHub/Kaggle_titanic/Data/test.csv')
test_data.shape

```

```python
test_data.isnull().sum()
```

```python
is_null = test_data['Age'].isnull()

test_data.loc[(test_data['Pclass'] == 1) & (test_data['Sex'] == "male")
    &(is_null), 'Age'] = float(41)
test_data.loc[(test_data['Pclass'] == 1) & (test_data['Sex'] == "female")
    &(is_null), 'Age'] = float(34)

test_data.loc[(test_data['Pclass'] == 2) & (test_data['Sex'] == "male")
    &(is_null), 'Age'] = float(30)
test_data.loc[(test_data['Pclass'] == 2) & (test_data['Sex'] == "female")
    &(is_null), 'Age'] = float(28)

test_data.loc[(test_data['Pclass'] == 3) & (test_data['Sex'] == "male")
    &(is_null), 'Age'] = float(26)
test_data.loc[(test_data['Pclass'] == 3) & (test_data['Sex'] == "female")
    &(is_null), 'Age'] = float(22)
```

```python
test_data['Cabin'] = test_data['Cabin'].apply(lambda x: 0 if pd.isnull(x) else 1)
```

```python
test_data.isnull().sum()
```

```python
test_data['Fare'] = test_data['Fare'].fillna(test_data['Fare'].mode()[0])
```

```python
test_data.isnull().sum()
```

```python
male = pd.get_dummies(test_data['Sex'], drop_first = True)
male
pd.get_dummies(test_data['Sex'])
```

```python
embarked = pd.get_dummies(test_data['Embarked'], drop_first = False)
embarked.head(3)
```

```python
test_data.head(10)
```

```python
xcol = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin']
x = test_data[xcol]
```

```python
x_temp = pd.concat([x,male], axis = 1)

x_temp.shape
```

```python
tmp = pd.concat([x_temp,embarked], axis = 1)
tmp.shape
```

```python
tmp.head(10)
```

```python
# Preprocess test data (apply the same preprocessing steps as used for training data)
# ...

# Load trained model
with open('/Users/masato/Documents/GitHub/Kaggle_titanic/Work/kaggle_survived.pkl', 'rb') as model_file:
    trained_model = pickle.load(model_file)

# Make predictions
predictions = trained_model.predict(tmp)

# Create prediction DataFrame
prediction_df = pd.DataFrame({
    'PassengerId': test_data['PassengerId'],
    'Survived': predictions
})

# Export predictions to CSV
prediction_df.to_csv('predictions.csv', index=False)
```
